```python
# Make function
facts = [i for i in data1.columns if 'ee_' in i or 'ep_' in i or 'incloss_' in i or 'cc_' in i]
agg_dict = {i: 'sum' for i in facts}

def eda_table0(data,field):
    a = data.groupby([field], dropna = False).agg(agg_dict).reset_index()
    
    for f in facts:
        a[f + '_pct'] = a[f]/a[f].sum()
        
    return a
```

```python
# Use function & change display settings
pd.set_option("display.max_columns",25)
pd.set_option("display.max_rows",250)
pd.options.display.float_format = '{:,.2f}'.format

from IPython.display import display, HTML

display(HTML(data="""
<style>
    div#notebook-container    { width: 90%; }
    div#menubar-container     { width: 65%; }
    div#maintoolbar-container { width: 99%; }
</style>
"""))

eda_table0(data1,'cbistier')
```

```python
# Make the function more specific
## % of exposures
## avg prem
## loss ratio
## freq, sev, pp

facts = [i for i in data1.columns if 'ee_' in i or 'ep_' in i or 'incloss_' in i or 'cc_' in i]
ee_list = [i for i in data1.columns if 'ee_' in i]
agg_dict = {i: 'sum' for i in facts}

covs = ['bi','col']

def eda_table(data,field):
    a = data.groupby([field], dropna = False).agg(agg_dict).reset_index()
    
    for e in ee_list:
        a[e + '_pct'] = a[e]/a[e].sum()
        
    for c in covs:
        a['avg_prem_' + c] = a['ep_' + c]/a['ee_' + c]
        a['lr_' + c] = a['incloss_' + c]/a['ep_' + c]
        a['freq_' + c] = a['cc_' + c]/a['ee_' + c]
        a['sev_' + c] = a['incloss_' + c]/a['cc_' + c]
        a['pp_' + c] = a['incloss_' + c]/a['ee_' + c]
        
    a.drop(columns = facts, inplace = True)  
#     a = a.drop(columns = facts)  
    
    return a
```

```python
# getting tired of repetitive tasks, let's make a function that creates a new field in the data with a suffix "_box" that will reduce the range of numeric fields
def box_var(data, field, low, high, new_name = False):
    if new_name == False:
        data[field] = np.where(data[field] <= low, low, np.where(data[field] >= high, high, data[field]))
    else:
        data[new_name] = np.where(data[field] <= low, low, np.where(data[field] >= high, high, data[field]))
    
box_var(data1,'numveh',1,4, new_name = 'veh_count_box')

remove_list.append('numveh')
stuff_to_analyze.remove('numveh')

eda_table(data1,'veh_count_box')
```

```python
# and which ones have decimal values?

float_list = []
for i in [i for i in data2.columns]:
    a = data2[([i])].drop_duplicates()
    a['rounded'] = round(a[i],0)
    a['rounded_diff'] = np.where(a['rounded'] != a[i],1 ,0)
    b = a.loc[a['rounded_diff'] == 1]
    
    if b.shape[0] > 0:
        print(i)
        print(b.head())
        
        float_list.append(i)
```

```python
# WHOOPS - there is a date in there which we can't round
# Let's try this again with some error handling

float_list = []
for i in [i for i in data2.columns]:
    try:
        a = data2[([i])].drop_duplicates()
        a['rounded'] = round(a[i],0)
        a['rounded_diff'] = np.where(a['rounded'] != a[i],1 ,0)
        b = a.loc[a['rounded_diff'] == 1]

        if b.shape[0] > 0:
            print(i)
            print(b.head())

            float_list.append(i)
    except:
        pass
```

```python
float_list
```

```python
# since all minimums besides the float_list items are non-negative, we can use uint instead of int

# max values by type:

# uint8, uint16, uint32
# 255,   65535, 4294967295

# let's make lists of fields based on max value and minimum datatype

remaining = [i for i in data2.columns if i not in float_list and i != 'pol_eff_date']
remaining
```

```python
uint8_list = []
uint16_list = []
uint32_list = []
error_list = []

for r in remaining:
    if data2[r].max() <= 255:
        uint8_list.append(r)
    elif (data2[r].max() <= 65535) & (data2[r].max() > 255):
        uint16_list.append(r)
    elif (data2[r].max() <= 4294967295) & (data2[r].max() > 65535):
        uint32_list.append(r)
    else:
        error_list.append(r)
        
print(uint8_list)
print(uint16_list)
print(uint32_list)
print(error_list)
```

```python
# make function to change datatype
# only need to do float, unit8, and uin16 because uint32_list and error_list are empty

print('Before: ' + str(round(data2.memory_usage(deep=True).sum()/1024**3,3)))

def cast_dtyp(data,field,dtyp):
    data[field] = data[field].astype(dtyp)
    
# now let's make a list of lists and the dtype we want to cast it to
mod_list = [float_list, uint8_list, uint16_list]
dtyp_list = ['float64','uint8','uint16']

for i, lst in enumerate(mod_list):
    for h in lst:
        cast_dtyp(data2,h,dtyp_list[i])
        
print('After: ' + str(round(data2.memory_usage(deep=True).sum()/1024**3,3)))
```
